```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows = 2, ncols = 3, tight_layout = True)

y = [1, -2, 5, 2, 2]
axes[0, 0].plot(y)

x = np.linspace(0, 10, 100)
y = 2 + 3 * np.sin(2 * x)
axes[0 , 1].plot(x, y)
axes[0, 1].set(xlim = (0, 10), xticks = np.arange(0,10),
               ylim = (-1, 5), yticks = np.arange(-2,7))

x = np.random.normal(20, 8 , 2000)
axes[0, 2].set(xlim = (-10, 50))
axes[0, 2].hist(x, bins = 30)

np.random.seed(3)
x = np.random.normal(6, 4, 60)
y = np.random.normal(4, 8, len(x))
axes[1,0].scatter(x, y)

plt.show()
```
